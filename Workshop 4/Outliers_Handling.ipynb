## **1. Introduction to Outliers**


### **What are Outliers?**
Outliers are values that are much higher or lower than most of the other values in a dataset. They are different from the rest of the data and can be considered unusual.

### Example:


Imagine a class of 100 students. If 98 students scored between 40 and 60 in a test, but 2 students scored 95 or 100, those 2 students are **outliers**. Their scores are much higher than the rest of the students, making them stand out as outliers in the data.

---

### **Effect of Outliers in Machine Learning**
- **Negative Effects**:
  - Skew the results of statistical analyses.
  - Affect the performance of machine learning models (e.g., linear regression, KNN).
  - Lead to incorrect conclusions or predictions.
- **Positive Effects**:
  - Can provide valuable insights (e.g., fraud detection, rare events).

---

### **How to Detect Outliers?**
| **Technique**            | **Description**                                                                 |
|--------------------------|---------------------------------------------------------------------------------|
| **Box Plot**             | Visualizes the distribution of data and identifies outliers using IQR.         |
| **Z-Score**              | Measures how many standard deviations a data point is from the mean.           |
| **IQR Method**           | Uses the Interquartile Range (IQR) to identify outliers.                       |
| **Scatter Plot**         | Visualizes outliers in bivariate data.                                         |
| **Histogram**            | Shows the distribution of data and highlights extreme values.                  |

---

### **How to Remove Outliers?**
| **Technique**            | **Description**                                                                 |
|--------------------------|---------------------------------------------------------------------------------|
| **Trimming**             | Removes outliers from the dataset.                                             |
| **Capping**              | Replaces outliers with a specified upper or lower limit.                       |
| **Transformation**       | Applies mathematical transformations (e.g., log, square root) to reduce skewness. |
| **Imputation**           | Replaces outliers with the mean, median, or mode.                              |

---

### **2. Outlier Removal Techniques**

 
### **1. Trimming**
#### **What is Trimming?**
Trimming involves **removing outliers** from the dataset entirely. This is done by identifying data points that fall outside a specified range (e.g., using the IQR method or Z-Score) and deleting them.

#### **Why Use Trimming?**
- Useful when outliers are clearly **errors** or **irrelevant** to the analysis.
- Helps in reducing the skewness of the data and improving the performance of machine learning models.

#### **Example Situation**:
Imagine you have a dataset of **student heights** in a school. Most students are between 4 and 6 feet tall, but there are a few entries with heights like 10 feet or 2 feet. These are likely **errors** and should be removed.

#### **How to Implement**:
1. Calculate the **Interquartile Range (IQR)**.
2. Define the lower and upper bounds:
   - Lower Bound = Q1 - 1.5 * IQR
   - Upper Bound = Q3 + 1.5 * IQR
3. Remove data points outside these bounds.

---

### **2. Capping**
#### **What is Capping?**
Capping involves **replacing outliers** with a specified upper or lower limit. Instead of removing the outliers, you set them to a maximum or minimum value.

#### **Why Use Capping?**
- Useful when outliers are **valid** but **extreme**.
- Preserves the data while reducing the impact of outliers.

#### **Example Situation**:
Consider a dataset of **employee salaries**. Most salaries range from $30,000 to $150,000, but there are a few outliers with salaries over $1,000,000. Instead of removing these outliers, you can **cap** them at the 99th percentile (e.g., $200,000).

#### **How to Implement**:
1. Calculate the **percentile values** (e.g., 1st and 99th percentiles).
2. Replace values below the lower percentile with the lower limit.
3. Replace values above the upper percentile with the upper limit.

---

### **3. Transformation**
#### **What is Transformation?**
Transformation involves applying **mathematical functions** (e.g., log, square root) to the data to reduce skewness and the impact of outliers.

#### **Why Use Transformation?**
- Useful when the data is **highly skewed**.
- Helps in making the data more **normally distributed**.

#### **Example Situation**:
Suppose you have a dataset of **house prices** where most houses are priced between $100,000 and $500,000, but there are a few mansions priced over $10,000,000. Applying a **log transformation** can reduce the impact of these extreme values.

#### **How to Implement**:
1. Apply a mathematical function (e.g., log, square root) to the data.
2. Ensure the transformation reduces skewness and makes the data more symmetric.

---

### **4. Imputation**
#### **What is Imputation?**
Imputation involves **replacing outliers** with a central tendency measure (e.g., mean, median, or mode).

#### **Why Use Imputation?**
- Useful when outliers are **errors** but the data is **valuable**.
- Helps in preserving the dataset size while reducing the impact of outliers.

#### **Example Situation**:
Imagine a dataset of **patient ages** where most patients are between 20 and 80 years old, but there are a few entries with ages like 200 or -10. These are likely **errors** and can be replaced with the **median age**.

#### **How to Implement**:
1. Calculate the **mean**, **median**, or **mode** of the column.
2. Replace outliers with the calculated value.

---

### **5. Winsorization**
#### **What is Winsorization?**
Winsorization is a combination of **capping and trimming**. It involves replacing outliers with the nearest non-outlier values.

#### **Why Use Winsorization?**
- Useful when you want to **preserve the data** but reduce the impact of extreme values.
- Less aggressive than trimming but more effective than capping.

#### **Example Situation**:
Consider a dataset of **test scores** where most scores range from 50 to 100, but there are a few scores below 20 or above 120. Winsorization replaces these extreme values with the nearest valid scores (e.g., 50 and 100).

#### **How to Implement**:
1. Define the lower and upper percentiles (e.g., 5th and 95th percentiles).
2. Replace values below the lower percentile with the lower percentile value.
3. Replace values above the upper percentile with the upper percentile value.

---

## **Summary Table of Techniques**

|---------------------|---------------------------------------------------------------------------------|-----------------------------------------------------------------------------|

### **Dataset 1: Iris Dataset (Box Plot + Trimming)**

```python

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris

# Load the Iris dataset
iris = load_iris()
df = pd.DataFrame(iris.data, columns=iris.feature_names)

# Detect Outliers using Box Plot
plt.figure(figsize=(10, 6))
sns.boxplot(data=df)
plt.title('Box Plot of Iris Dataset')
plt.show()

# Remove Outliers using IQR
Q1 = df.quantile(0.25)
Q3 = df.quantile(0.75)
IQR = Q3 - Q1
df_trimmed = df[~((df < (Q1 - 1.5 * IQR)) | (df > (Q3 + 1.5 * IQR))).any(axis=1)]

print("Original Dataset Shape:", df.shape)
print("Trimmed Dataset Shape:", df_trimmed.shape)
```

```python
df.head()
```

```python
# Detect Outliers using Box Plot
plt.figure(figsize=(10, 6))
sns.boxplot(data=df_trimmed)
plt.title('Box Plot of Iris Dataset')
plt.show()
```


**Explanation**:  
- Outliers are detected using a box plot and removed using the IQR method.
- Used because the Iris dataset has some extreme values.

### **Dataset 2: Titanic Dataset (Z-Score + Capping)**

![](https://i.ytimg.com/vi/Zoi15vIgDQ8/maxresdefault.jpg)

```python

import pandas as pd
import numpy as np
#!pip install scipy
from scipy.stats import zscore
#z score = score - mean(score) / std(score)

# Load the Titanic dataset
url = "https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv"
df = pd.read_csv(url)

# Detect Outliers using Z-Score
df['zscore'] = zscore(df['Fare'])

outliers = df[(df['zscore'] > 3) | (df['zscore'] < -3)]

# Cap Outliers

df['Fare_capped'] = np.where(df['zscore'] > 3, df['Fare'].quantile(0.99), df['Fare'])

print("Outliers in Fare Column:\n", outliers[['Fare', 'zscore']])
```

**Explanation**:  
- Outliers in the `Fare` column are detected using Z-Score and capped at the 99th percentile.
- Used because the `Fare` column has extreme values.

### **Dataset 3: Wine Quality Dataset (IQR Method + Transformation)**

```python

import pandas as pd
import numpy as np

# Load the Wine Quality dataset
url = "https://archive.ics.uci.edu/ml/machine-learning-databases/wine-quality/winequality-red.csv"
df = pd.read_csv(url, delimiter=';')

# Detect Outliers using IQR
Q1 = df['alcohol'].quantile(0.25)
Q3 = df['alcohol'].quantile(0.75)
IQR = Q3 - Q1
outliers = df[(df['alcohol'] < (Q1 - 1.5 * IQR)) | (df['alcohol'] > (Q3 + 1.5 * IQR))]

# Apply Log Transformation
df['alcohol_log'] = np.log(df['alcohol'])

print("Outliers in Alcohol Column:\n", outliers[['alcohol']])
```

```python
df.head()
```

**Explanation**:  
- Outliers in the `alcohol` column are detected using the IQR method and reduced using a log transformation.
- Used because the `alcohol` column is highly skewed.

### **Dataset 4: Cars Dataset (Scatter Plot + Imputation)**

```python

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the Cars dataset
url = "https://raw.githubusercontent.com/selva86/datasets/master/Cars93_miss.csv"
df = pd.read_csv(url)

# Detect Outliers using Scatter Plot
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Horsepower', y='Price', data=df)
plt.title('Scatter Plot of Horsepower vs Price')
plt.show()

# Impute Outliers with Median
median_price = df['Price'].median()
df['Price_imputed'] = np.where(df['Price'] > 50, median_price, df['Price'])

print("Outliers in Price Column:\n", df[df['Price'] > 50][['Price']])
```

```python
df.head()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Horsepower', y='Price_imputed', data=df)
plt.title('Scatter Plot of Horsepower vs Price')
plt.show()
```

**Explanation**:  
- Outliers in the `Price` column are detected using a scatter plot and replaced with the median value.
- Used because the `Price` column has extreme values.

### **Dataset 5: Housing Dataset (Histogram + Capping)**


```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# Load the Housing dataset
url = "https://raw.githubusercontent.com/ageron/handson-ml2/master/datasets/housing/housing.csv"
df = pd.read_csv(url)

# Plot original distribution
plt.figure(figsize=(10, 6))
sns.histplot(data=df[df['median_house_value'] <= 490000], x='median_house_value', bins=50)
plt.title('Distribution of Median House Value (Excluding Values > 490000)')
plt.show()

# Cap values above 490000
threshold = 490000
df['median_house_value_capped'] = df['median_house_value'].clip(upper=threshold)

# Show number of capped values
print("Number of values capped (above 490000):")
print(len(df[df['median_house_value'] > threshold]))

# Plot distribution after capping
plt.figure(figsize=(10, 6))
sns.histplot(data=df, x='median_house_value_capped', bins=50)
plt.title('Distribution of Median House Value (After Capping at 490000)')
plt.show()

# Compare statistics
print("\nBefore vs After Capping Statistics:")
print(pd.DataFrame({
    'Original': df['median_house_value'].describe(),
    'Capped': df['median_house_value_capped'].describe()
}))
```

**Explanation**:  
- Outliers in the `median_house_value` column are detected using a histogram and capped at 500,000.
- Used because the `median_house_value` column has extreme values.

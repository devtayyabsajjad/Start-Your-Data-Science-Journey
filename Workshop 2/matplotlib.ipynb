```python

#Bar Chart 


labels=["A","B","C","D","E"]
values=[10,20,30,40,50]

palletes=["red","blue","green","yellow","purple"]

a=plt.bar(labels,values,color=palletes)

pattern=["/","*","o","|","."]
for i in range(len(a)):
    a[i].set_hatch(pattern[i])




plt.title("Bar Chart")
plt.xlabel("Labels")
plt.ylabel("Values")
plt.show()

```

# **Fifa Dataset**


```python
df=pd.read_csv("fifa_data.csv")

#set the head to see all columns
pd.set_option("display.max_columns",None)
df.head()   



```

```python
df.columns
```

```python

import warnings
warnings.filterwarnings("ignore")
# Pie Chart 


left_foot=df[df["Preferred Foot"]=="Left"].count()[0]
right_foot=df[df["Preferred Foot"]=="Right"].count()[0]

#PIE CHART ON LEFT AND RIGHT FOOT
plt.figure(figsize=(8,6),dpi=240)

plt.pie([left_foot,right_foot],labels=["Left","Right"],explode=[0.1,0],colors=["orange","lightblue"],autopct="%1.1f%%")
plt.title("Left and Right Foot")
plt.show()


```

```python


barcolona=df.loc[df.Club=="FC Barcelona"]["Overall"]
real_madrid=df.loc[df.Club=="Real Madrid"]["Overall"]
newcastle=df.loc[df.Club=="Newcastle United"]["Overall"]
liverpool=df.loc[df.Club=="Liverpool"]["Overall"]



#Boxplot on overall score for club
labels=["FC Barcelona","Real Madrid","Newcastle United","Liverpool"]
plt.boxplot([barcolona,real_madrid,newcastle,liverpool],labels=labels)
plt.title("Overall Score for Club")
plt.xlabel("Club")
plt.ylabel("Overall Score")
plt.show()


#Bar chart on overall score for club

```

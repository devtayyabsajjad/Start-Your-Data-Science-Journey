# Categorical Data
## 1.Countplot

```python
sns.countplot(x='Tutoring_Sessions',data=df)
```

```python
sns.countplot(x='Gender', data=df)
```

```python
sns.countplot(x='Access_to_Resources',hue='Gender', data=df)
plt.title("Access to Resources")
plt.xlabel('(Students)')
plt.ylabel(('Numbers'))
```

## Pie Chart

```python
df['Gender'].value_counts().plot(kind='pie', autopct='%1.1f%%', startangle=90, colors=['#66b3ff', '#99ff99'])
```

```python
df['Parental_Involvement'].value_counts().plot(kind='pie', autopct='%1.1f%%', startangle=90)
plt.title("Access to Resources")
```

## Barplot (Numerical - Categorical)

```python
pivot_data = df.pivot_table(index='Parental_Involvement', columns='Gender', values='Exam_Score', aggfunc='mean')
pivot_data.plot(kind='bar', stacked=True, figsize=(10, 6))
```

```python
# Box Plot
sns.boxplot(x='Internet_Access', y='Exam_Score', data=df)
```

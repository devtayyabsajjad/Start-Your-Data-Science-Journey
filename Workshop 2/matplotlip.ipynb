## Why Use Matplotlib?

- **Versatility:** You can create line charts, bar charts, scatter plots, histograms, and more.

- **Ease of Use:** Simple commands can generate powerful visualizations.
- **Compatibility:** It integrates well with other libraries like NumPy, Pandas, and Seaborn.

## Step 1: Import Matplotlib


`pyplot` is the submodule of Matplotlib that provides a MATLAB-like interface for creating visualizations.

---

```python
import matplotlib.pyplot as plt
```


1. `plt.figure()`: Creates## a new drawing board
2. `plt.plot()`: Draws lines
3. `plt.bar()`: Creates bar charts
4. `plt.scatter()`: Makes dot plots
5. `plt.pie()`: Shows proportions
6. `plt.hist()`: Displays data distribution

## Step 2: Creating Common Charts & Plot
 
### Letâ€™s start with a **line plot**, one of the most basic types of plots.
### Our First Plot: Simple Line ðŸ“ˆ


```python
# Example: Line Plot
import matplotlib.pyplot as plt

# Data
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

# Create the plot
plt.plot(x, y)

# Add labels and title
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Basic Line Plot")

# Show the plot
plt.show()
```

## Step 2.1 : Customize Your Line Chart

Letâ€™s make the chart more informative by adding:
- **Grid lines**
- **Markers** to highlight the points
- **A different color** for the line


```python
# Data for the line chart
x = [1, 2, 3, 4, 5]
y = [1, 4, 9, 16, 25]

# Creating the plot
plt.plot(x, y, color='blue', marker='o', linestyle='--')  # Adding color, markers, and style

# Adding labels and title
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Customized Line Chart")

# Adding grid lines
plt.grid(True)

# Showing the plot
plt.show()
```

### Step 2.2: Adding Multiple Lines to the Same Chart

Sometimes, we want to compare two datasets on the same chart. Letâ€™s add another line.

### Code Example

```python
# Data for two lines
x = [1, 2, 3, 4, 5]
y1 = [1, 4, 9, 16, 25]  # First line
y2 = [1, 8, 27, 64, 125]  # Second line

# Plotting the first line
plt.plot(x, y1, color='blue', marker='o', label='y = x^2')

# Plotting the second line
plt.plot(x, y2, color='red', marker='s', label='y = x^3')

# Adding labels and title
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Multiple Lines Chart")

# Adding a legend
plt.legend()

# Showing the plot
plt.show()
```

## 3: Bar Chart: Comparing Things ðŸ“Š
#### Imagine you want to compare how much ice cream different friends ate


```python
# Example: Bar Chart
categories = ["A", "B", "C", "D"]
values = [4, 7, 1, 8]

plt.bar(categories, values, color='skyblue')
plt.title("Bar Chart Example")
plt.xlabel("Categories")
plt.ylabel("Values")
plt.show()
```

### 3.2 Scatter Plot: Finding Patterns ðŸ•µï¸
#### Like throwing colored dots on a map to see if they make a pattern




```python
# Example: Scatter Plot
x = [5, 7, 8, 7, 2, 17, 2, 9, 4, 11]
y = [99, 86, 87, 88, 100, 86, 103, 87, 94, 78]

plt.scatter(x, y, color='red')
plt.title("Scatter Plot Example")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

### 3.3 Histogram: Seeing Data Distribution ðŸ“ˆ
#### Like sorting marbles into boxes to see how they're spread out



```python
# Example: Histogram
data = [22, 87, 5, 43, 56, 73, 55, 54, 11, 20, 51, 5, 79, 31, 27]

plt.hist(data, bins=5, color='green', edgecolor='black')
plt.title("Histogram Example")
plt.xlabel("Bins")
plt.ylabel("Frequency")
plt.show()
```

### 3.4 Pie Chart Showing Proportions ðŸ•
#### Like cutting a pizza and seeing how big each slice is

---


```python
# Example: Pie Chart
labels = ['Python', 'Java', 'C++', 'Ruby']
sizes = [215, 130, 245, 210]
colors = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue']

explode = (0.1, 0, 0, 0)  # explode 1st slice

plt.pie(sizes, explode=explode, labels=labels, colors=colors, autopct='%1.1f%%', shadow=True, startangle=140)
plt.title("Pie Chart Example")
plt.axis('equal')
plt.show()
```

## **1. What is Seaborn?** 🐧
Seaborn is a Python data visualization library built on top of Matplotlib. It provides a high-level interface for creating attractive and informative statistical graphics with ease.

Key Features:
- Beautiful default styles
- Simple syntax for complex plots
- Integrated with Pandas for easy data handling with Pandas and NumPy

Seaborn simplifies the creation of advanced visualizations. It handles many repetitive tasks, such as aggregating data and applying statistical transformations, making it an excellent choice for exploratory data analysis.

**Advantages of Seaborn:**
- Concise and intuitive syntax
- Automatic estimation of data relationships
- Built-in themes for beautiful plots

## **Bar Plot** 📊
Bar plots represent data using rectangular bars.

- **What**: A plot showing the relationship between categorical and numerical data using bars.
- **Why**: Used to compare quantities across categories.
- **Data Type**: Categorical (x-axis) and numerical (y-axis).

```python
# 7. Barplot
plt.figure(figsize=(12, 8), dpi=200)
sns.barplot(x="day", y="total_bill", data=data2)
plt.show()
```

```python

# 7. Barplot

custom_palette = [   "#F7FF33", "#FF33D4"]

sns.barplot(x="day", y="total_bill",  hue="sex",     data=data2, palette=custom_palette)


# how can we add title in our Seaborn charts 

plt.title("Barplot: Total Bill by Day and Gender")


plt.show()
```

```python

# 7. Barplot
sns.barplot(x="day", y="total_bill", hue="sex", data=data2, color = "green" )

plt.title("Barplot: Total Bill by Day and Gender")

plt.show()
```

```python
# Define a custom color palette with multiple colors
custom_palette = [  "#3357FF", "#F7FF33", "#FF33D4"]

sns.barplot(x="day", y="total_bill", hue="sex", data=data2,palette=custom_palette )
plt.title("Barplot: Total Bill by Day and Gender")
plt.show()
```

```python

# Horizontal Barplot
sns.barplot(y="day", x="total_bill", hue="sex", data=data2, palette="husl")

plt.title("Horizontal Barplot: Total Bill by Day and Gender")
plt.show()
```

## **Box Plot** 📦
Box plots visualize the spread and outliers in the data.

- **What**: Visualizes the distribution of data through their quartiles and highlights outliers.
- **Why**: Used to compare distributions across categories and detect outliers.
- **Data Type**: Numerical data with optional categorical grouping.


![](https://www.simplypsychology.org/wp-content/uploads/box-whisker-plot.jpg)



```python
# 8. Box Plot
plt.figure(figsize=(8, 6))

sns.boxplot(x="species", y="petal_width", data=data1)
plt.title("Box Plot: Petal Width by Species")
plt.show()
```

## **Joint Grid** 🧑‍🤝‍🧑
JointGrid allows advanced customization for paired plots.

- **What**: A grid-based plot for visualizing the relationship between two variables along with their marginal distributions.
- **Why**: Used to explore bivariate relationships and individual distributions simultaneously.
- **Data Type**: Bivariate numerical data.

```python
# Joint Grid
g = sns.JointGrid(data=data2, x="total_bill", y="tip")
g.plot(sns.scatterplot, sns.histplot)
plt.title("JointGrid: Total Bill vs Tip")
plt.show()
```

```python
# Joint Plot
sns.jointplot(data=data2, x="total_bill", y="tip", kind="reg")
plt.suptitle("Joint Plot with Regression: Total Bill vs Tip", y=1.02)
plt.show()
```

## **Dis Plot** 📊
Distribution plots show the histogram of data along with optional KDE.

- **What**: A distribution plot showing histograms or KDE plots.
- **Why**: Used to visualize the distribution of a single numerical variable.
- **Data Type**: Univariate numerical data.

```python
# 10. Dis Plot

sns.displot(data=data2, x="total_bill", kde=True, bins=20)
plt.title("Displot: Total Bill Distribution")
plt.show()
```

## **Heatmap** 🌡️
A heatmap is used to visualize data in a matrix format with colors representing values.

- **What**: A graphical representation of data using a matrix where individual values are represented as colors.
- **Why**: Used for visualizing relationships between variables and identifying patterns or anomalies.
- **Data Type**: Numerical (e.g., correlation matrix).
---

```python


plt.figure(figsize=(16, 8))

corr = data1.select_dtypes(include=['float64', 'int']).corr()  # Correlation matrix for numeric columns
sns.heatmap(corr, annot=True, cmap="coolwarm")
plt.title("Heatmap of Iris Dataset")
plt.show()
```

## **Violin Plot** 🎻
Violin plots are used to visualize the distribution of numeric data.

- **What**: Combines a box plot and a KDE plot to show the distribution and probability density of data.
- **Why**: Used to visualize the distribution of numerical data across different categories.
- **Data Type**: Numerical data with categorical grouping.


```python
#  Violin Plot
sns.violinplot(x="species", y="sepal_length", data=data1)
plt.title("Violin Plot: Sepal Length by Species")
plt.show()
```

## **Pair Plot** 🌀
Pair plots show pairwise relationships across the dataset.

- **What**: Displays pairwise relationships in a dataset using scatterplots and histograms/KDE plots for marginal distributions.
- **Why**: Used for exploratory data analysis to identify relationships or clusters.
- **Data Type**: Multivariate numerical data.


```python
# Pairplot
sns.pairplot(data=data1, hue="species")
plt.suptitle("Pairplot of Iris Dataset", y=1.02)
plt.show()
```
